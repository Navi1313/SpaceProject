## 2) Model Parameters and HyperParameters

### Insted of using an gradient descent and y-intercept, we will be using the actual model parameters for ML algo - Weight and bais 

```python
# Initialize the weight (w)  and bais (b) to 0 :
W = 0 
b = 0

# Hyper Parameters -> Choose appropriate LRate (0.01)
LRate  = 0.01 
```

```python
# Create a function for Predicted Output :->
def predict(X , W , b):
    '''
    X -  X array data 
    W - weight (m)
    b - biased value (c : intercept)
    '''
    return W*X + b 
```

```python
# Creating an MSE cost function :->

def cost_fun(X , y, W , b ):
    '''
    X array data 
    W - weight (m)
    b - biased value (c : intercept)
    y - predicted value of y : 
    '''
    y_pred = predict(X , W , b)
    squaredError = (y-y_pred)**2
    
    MSE = np.mean(squaredError)
    return MSE 
```

```python
cost_fun(x_train , y_train , w , b )
```

<h1 style="text-align: center; color: yellow;"> <i>Backward Propogation Equation - Linear Regression  <i> </h1>

```python
# Creating an function to update an parameters 

def update_params(X , y, W , b , LRate ):
    '''
    X array data 
    W - weight (m)
    b - biased value (c : intercept)
    y - predicted value of y : 

    Return - tuple with biased term (b) and Weight (W)
    '''
    y_p  =  predict(x_train , W , b)
    dldW =  np.mean((-2 * X) * ( y - y_p ))
    dldb =  np.mean(-2 * (y - y_p))
    W    =   W - LRate * dldW
    b    =   b - LRate * dldb

    return W ,b 
```

```python
update_params(x_train , y_train, W , b , LRate )
```

```python
def training(X, y, W, b, LRate, thold=1e-20, verbose=False):
    '''
    X        - array data 
    W        - weight (m)
    b        - biased value (c : intercept)
    y        - predicted value of y 
    thold    - Threshold for training should stop based on convergence 
    verbose  - display the training if it is set to True 

    Return:
    -----------

    weights  - list of all the updated values in the iterations (epochs)
    biases   - list of all the values updated in iterations 
    cost     - list of all the cost values based on W and b values in iteration 
    itr_count - iteration count at the end of training 
    '''

    # 1) Initialize counter
    itr_count =  0 

    # Lists to store updated weights, biases, and costs
    weights = [W]
    biases = [b]
    costs = [cost_fun(X, y, W, b)]

    '''
    This is the Process we need to follow ->
    
        0) Initialize counter ..
        1) Calculate the current value of weight and bias.
        2) Add this cost to the list of costs.
        3) Break the training loop if the cost value is converging.
        4) If not, update values of W and b.
        5) Add these values to the respective list. 
        6) Print the training progress after every 100 iterations.
        7) Update the counter value 
    '''

    while True:
        # 2) Calculate the cost and append into the list         
        cost = cost_fun(X, y, W, b)
        costs.append(cost)

        # 3) Break the training loop if the cost value is converging         
        if itr_count > 2 and abs(costs[-1] - costs[-2]) < thold:
            break
            
        # 4) If not, update values of W and b
        W, b = update_params(X, y, W, b, LRate)
        
        # 5) Add these values to the respective list 
        weights.append(W)
        biases.append(b)

        # 6) Print the training progress after every 100 iterations  
        if verbose and (itr_count + 1) % 100 == 0:
            print(f"After {itr_count + 1} iterations   :->   weight= {W}, biases= {b}, cost= {cost}")

        # 7) Update the counter value
        itr_count += 1 

    return weights, biases, costs, itr_count
```

```python
weights , biases , costs , epoachs  = training(x_train, y_train, W, b, LRate, thold=1e-13, verbose=True)
```

```python
# Comparing initial AND Final values 

import pandas as pd 

inital = [weights[0] , biases[0] , costs[0]]
final =  [weights[-1] , biases[-1] , costs[-1]]

df = pd.DataFrame(list(zip(inital , final)), columns=['initial' , 'Final'], index = ['weight', 'Bais', 'Cost'])
df
```

```python
# Visualizing the cost wrt epochs --> Figuresize (15 ,6) , epoach on x and cost on y axis 

plt.figure(figsize= (15,6))
plt.plot(costs)
plt.xlabel("Epoachs")
plt.ylabel("Cost")
plt.title(f"Display a Cost change vs epoachs (Total no of : {epoachs})")
plt.show()

```

```python
# DeFINE tHE SEED value (diffrent seed value than an training dataset avoid leakage ) 
np.random.seed(5007)

# Test Data
x_test = 3 * np.random.rand(10 ,1)
y_test = 2 * x_test  +  9  +  np.random.rand(10 , 1)

prediction = predict(x_test , weights[-1] , biases[-1])
```
